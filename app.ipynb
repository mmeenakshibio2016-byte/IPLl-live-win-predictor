```python
required_columns = [
    'match_id',
    'innings',
    'batting_team',
    'bowling_team',
    'venue',
    'team_runs',
    'team_wicket',
    'team_balls',
    'runs_target',
    'match_won_by'
]

df = df[required_columns]
```

```python
# TARGET: Team win (Corrected 'winner' to 'match_won_by')
df.loc[:, 'win'] = (df['batting_team'] == df['match_won_by']).astype(int)


df = df[df['innings'] == 2].copy()

# FEATURE ENGINEERING - Use existing cumulative columns directly
df.loc[:, 'current_score'] = df['team_runs']
df.loc[:, 'wickets_lost'] = df['team_wicket']
df.loc[:, 'balls_bowled'] = df['team_balls']

# Ensure 'balls_bowled' is greater than 0 before proceeding
df = df[df['balls_bowled'] > 0].copy()

# Forward-fill 'runs_target' within each match_id, then drop any remaining NaNs
# This assumes runs_target is set once per match and should apply to all subsequent balls in the innings
df.loc[:, 'runs_target'] = df.groupby('match_id')['runs_target'].ffill()
# Drop rows where runs_target is still NaN (e.g., no target was ever set in the match)
df = df.dropna(subset=['runs_target']).copy()

# Calculate chase-specific metrics (Corrected 'target' to 'runs_target')
df.loc[:, 'runs_left'] = df['runs_target'] - df['current_score']
df.loc[:, 'balls_left'] = 120 - df['balls_bowled']
df.loc[:, 'wickets_left'] = 10 - df['wickets_lost']

# Handle potential division by zero for run rates
df.loc[:, 'current_run_rate'] = df['current_score'] * 6 / df['balls_bowled']
df.loc[:, 'required_run_rate'] = df['runs_left'] * 6 / df['balls_left']

# Replace inf/-inf resulting from division by zero with NaN, then drop those rows
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df = df.dropna(subset=['current_run_rate', 'required_run_rate']).copy()

# REMOVE END-OF-MATCH LEAKAGE and apply final filters for a valid chase: still balls and runs left
df = df[(df['balls_left'] > 0) & (df['runs_left'] > 0)].copy()
```

```python
features = [
    'batting_team',
    'bowling_team',
    'venue',
    'runs_left',
    'balls_left',
    'wickets_left',
    'current_run_rate',
    'required_run_rate',
    'innings'
]

# Create the 'win' column: 1 if batting_team won, 0 otherwise
df.loc[:, 'win'] = (df['batting_team'] == df['match_won_by']).astype(int)

df_final = df[['match_id'] + features + ['win']].dropna()
```

```python
from sklearn.model_selection import train_test_split

match_ids = df_final['match_id'].unique()

train_ids, test_ids = train_test_split(
    match_ids,
    test_size=0.3,
    random_state=42,

)

train_df = df_final[df_final['match_id'].isin(train_ids)]
test_df = df_final[df_final['match_id'].isin(test_ids)]

X_train = train_df.drop(['match_id', 'win'], axis=1)
y_train = train_df['win']

X_test = test_df.drop(['match_id', 'win'], axis=1)
y_test = test_df['win']
```

```python
# Identify categorical columns in the new X_train
categorical_cols = X_train.select_dtypes(include='object').columns

# Apply one-hot encoding to both training and testing sets
X_train_encoded = pd.get_dummies(X_train, columns=categorical_cols, drop_first=True)
X_test_encoded = pd.get_dummies(X_test, columns=categorical_cols, drop_first=True)

# Align columns - crucial for train and test sets to have the same columns after one-hot encoding
# This handles cases where a category might appear in train but not test, or vice versa
common_cols = list(set(X_train_encoded.columns) & set(X_test_encoded.columns))
X_train_final = X_train_encoded[common_cols]
X_test_final = X_test_encoded[common_cols]

```

```python
import streamlit as st
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score

st.set_page_config(page_title="IPL Live Win Predictor", layout="centered")

st.title("🏏 IPL Live Win Probability Predictor")

# -------------------------------
# LOAD DATA (SMALL CSV ONLY)
# -------------------------------
@st.cache_data
def load_data():
    df = pd.read_csv("")   # small dataset only
    return df

# -------------------------------
# TRAIN MODEL
# -------------------------------
@st.cache_resource
def train_model(df):
    features = [
        'batting_team',
        'bowling_team',
        'venue',
        'runs_left',
        'balls_left',
        'wickets_left',
        'current_run_rate',
        'required_run_rate'
    ]

    X = df[features]
    y = df['win']

    categorical = ['batting_team', 'bowling_team', 'venue']
    numerical = list(set(features) - set(categorical))

    encoder = OneHotEncoder(handle_unknown="ignore", sparse=False)
    X_cat = encoder.fit_transform(X[categorical])
    X_num = X[numerical].values

    X_final = np.hstack((X_cat, X_num))

    X_train, X_test, y_train, y_test = train_test_split(
        X_final, y, test_size=0.2, random_state=42
    )

    model = xgb.XGBClassifier(
        n_estimators=200,
        max_depth=6,
        learning_rate=0.05,
        eval_metric="logloss"
    )
    model.fit(X_train, y_train)

    return model, encoder, categorical, numerical

# -------------------------------
# LOAD + TRAIN
# -------------------------------
df = load_data()
model, encoder, cat_cols, num_cols = train_model(df)

# -------------------------------
# USER INPUTS
# -------------------------------
st.subheader("Match Situation")

batting_team = st.selectbox("Batting Team", sorted(df['batting_team'].unique()))
bowling_team = st.selectbox("Bowling Team", sorted(df['bowling_team'].unique()))
venue = st.selectbox("Venue", sorted(df['venue'].unique()))

runs_left = st.number_input("Runs Left", 0, 300, 60)
balls_left = st.number_input("Balls Left", 1, 120, 36)
wickets_left = st.number_input("Wickets Left", 0, 10, 6)

current_rr = st.number_input("Current Run Rate", 0.0, 20.0, 7.5)
required_rr = st.number_input("Required Run Rate", 0.0, 20.0, 10.0)

# -------------------------------
# PREDICTION
# -------------------------------
if st.button("Predict Winner"):
    input_df = pd.DataFrame([{
        'batting_team': batting_team,
        'bowling_team': bowling_team,
        'venue': venue,
        'runs_left': runs_left,
        'balls_left': balls_left,
        'wickets_left': wickets_left,
        'current_run_rate': current_rr,
        'required_run_rate': required_rr
    }])

    X_cat = encoder.transform(input_df[cat_cols])
    X_num = input_df[num_cols].values
    X_final = np.hstack((X_cat, X_num))

    win_prob = model.predict_proba(X_final)[0][1] * 100

    if win_prob >= 50:
        st.success(f"🏆 {batting_team} likely to WIN")
    else:
        st.warning(f"🏆 {bowling_team} likely to WIN")

    st.metric("Win Probability", f"{win_prob:.2f}%")
```
